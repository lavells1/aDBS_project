# aDBS sensing + stim settings - whole cohort

Iterates every patient under `DATA_ROOT` and runs the per-patient detection from
**`adbs_extract`** (the same logic as cell 7 of `adaptpd_sensing_waveform_viewer.ipynb`),
collecting one row **per patient x hemisphere** into a single `cohort_df`.

- `cohort_df` - full table (sensing channel + all stim/phase settings + lead target)
- `failures_df` - patients that could not be auto-detected (e.g. only 1 export, no ADBS_SETUP/eval/baseline)
- writes `adbs_cohort_sensing_stim.csv`

```python
from pathlib import Path
import importlib

import pandas as pd

import adbs_extract
importlib.reload(adbs_extract)  # pick up edits to the shared module without restarting the kernel

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 240)

DATA_ROOT = Path('/project/hammer_neuromod/RawData/ADAPT_PD/ADAPT-PD_Trial_JSONs')
patients = adbs_extract.list_patients(DATA_ROOT)
print(f'{len(patients)} patients found under {DATA_ROOT}')
```

```python
cohort_df, failures_df = adbs_extract.build_cohort(DATA_ROOT, patients)
n_pat = cohort_df['patient'].nunique() if not cohort_df.empty else 0
print(f'{n_pat} patient(s) extracted -> {len(cohort_df)} hemisphere-row(s); '
      f'{len(failures_df)} patient(s) skipped.')
cohort_df
```

```python
# Compact per-hemisphere summary (the most-used columns)
SUMMARY_COLS = [
    'patient', 'hemisphere', 'stim_target', 'waveform_channel', 'mode',
    'rate_hz', 'pulse_width_us', 'amp_lower_ma', 'amp_upper_ma', 'amp_suspend_ma',
    'lfp_lower_threshold', 'lfp_upper_threshold', 'lfp_beta_center_hz',
    'adaptive_status', 'group_name', 'n_files',
]
cohort_summary = cohort_df[SUMMARY_COLS] if not cohort_df.empty else cohort_df
cohort_summary
```
