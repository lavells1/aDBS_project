# aDBS programming over time - Visits 3 onwards (whole cohort)

Longitudinal companion to `adbs_cohort_sensing.ipynb`. For every patient, pulls the
adaptive program's sensing channel + stim/phase settings from each **VISIT_3 and later**
export (EXTEND_ACCESS included), giving **one row per patient x hemisphere x visit** so
you can see how programming evolved during the chronic phase.

- `visits_df` - full longitudinal table
- `failures_df` - patients with no Visit>=3 / EXTEND_ACCESS export (or no adaptive program in them)
- writes `adbs_cohort_visits_stim.csv`

```python
from pathlib import Path
import importlib

import pandas as pd

import adbs_extract
importlib.reload(adbs_extract)  # pick up edits to the shared module without restarting the kernel

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 240)

DATA_ROOT = Path('/project/hammer_neuromod/RawData/ADAPT_PD/ADAPT-PD_Trial_JSONs')
INCLUDE_EXTEND_ACCESS = True
patients = adbs_extract.list_patients(DATA_ROOT)
print(f'{len(patients)} patients found under {DATA_ROOT}')
```

```python
visits_df, failures_df = adbs_extract.build_cohort_visits(
    DATA_ROOT, patients, include_extend=INCLUDE_EXTEND_ACCESS
)
n_pat = visits_df['patient'].nunique() if not visits_df.empty else 0
print(f'{n_pat} patient(s) with Visit>=3 programming -> {len(visits_df)} row(s) '
      f'(patient x hemisphere x visit); {len(failures_df)} patient(s) skipped.')
visits_df.head(20)
```

```python
# Compact longitudinal summary (most-used columns)
SUMMARY_COLS = [
    'patient', 'visit', 'visit_number', 'timestamp', 'hemisphere', 'stim_target',
    'waveform_channel', 'mode', 'rate_hz', 'pulse_width_us',
    'amp_lower_ma', 'amp_upper_ma', 'amp_suspend_ma',
    'lfp_lower_threshold', 'lfp_upper_threshold', 'lfp_beta_center_hz',
    'adaptive_status', 'group_name',
]
visits_summary = visits_df[SUMMARY_COLS] if not visits_df.empty else visits_df
visits_summary
```
