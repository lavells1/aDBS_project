# aDBS programming over time - ALL visits (whole cohort)

Full longitudinal programming timeline. For every patient, pulls the adaptive program's
sensing channel + stim/phase settings from **every export that contains an adaptive
program** (consent -> screening -> baseline -> setup -> eval -> all VISITs -> EXTEND_ACCESS),
giving **one row per patient x hemisphere x export**.

Same shape as `adbs_cohort_visits.ipynb`, just without the Visit>=3 restriction
(`visit_filter=None`). Exports with no adaptive program contribute no rows.

- `timeline_df` - full longitudinal table
- `failures_df` - patients with no adaptive program in any export
- writes `adbs_cohort_all_visits_stim.csv`

```python
from pathlib import Path
import importlib

import pandas as pd

import adbs_extract
importlib.reload(adbs_extract)  # pick up edits to the shared module without restarting the kernel

pd.set_option('display.max_columns', None)
pd.set_option('display.width', 240)

DATA_ROOT = Path('/project/hammer_neuromod/RawData/ADAPT_PD/ADAPT-PD_Trial_JSONs')
patients = adbs_extract.list_patients(DATA_ROOT)
print(f'{len(patients)} patients found under {DATA_ROOT}')
```

```python
# visit_filter=None -> every export (full timeline)
timeline_df, failures_df = adbs_extract.build_cohort_timeline(DATA_ROOT, patients, visit_filter=None)
n_pat = timeline_df['patient'].nunique() if not timeline_df.empty else 0
print(f'{n_pat} patient(s) -> {len(timeline_df)} row(s) (patient x hemisphere x export); '
      f'{len(failures_df)} patient(s) skipped.')
timeline_df.head(20)
```

```python
# Compact longitudinal summary (most-used columns)
SUMMARY_COLS = [
    'patient', 'visit', 'visit_number', 'timestamp', 'hemisphere', 'stim_target',
    'waveform_channel', 'mode', 'rate_hz', 'pulse_width_us',
    'amp_lower_ma', 'amp_upper_ma', 'amp_suspend_ma',
    'lfp_lower_threshold', 'lfp_upper_threshold', 'lfp_beta_center_hz',
    'adaptive_status', 'group_name',
]
timeline_summary = timeline_df[SUMMARY_COLS] if not timeline_df.empty else timeline_df
timeline_summary
```
